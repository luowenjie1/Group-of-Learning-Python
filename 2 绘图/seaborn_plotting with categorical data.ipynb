```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from scipy import stats

import seaborn as sns
sns.set(style="ticks", color_codes=True)
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
tips.info()
```

```python
sns.catplot(x="day", y="total_bill", data=tips)#kind="strip"; the default
sns.catplot(x="day", y="total_bill", jitter=False, data=tips)
```

```python
print("swarm")
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips)
```

```python
print("box")
sns.catplot(x="day", y="total_bill", kind="box", data=tips) 
```

```python
print("violin")
sns.catplot(x="day", y="total_bill", kind="violin", data=tips) 
```

```python
print("boxen")
sns.catplot(x="day", y="total_bill", kind="boxen", data=tips)
```

# seaborn.pointplot
Show point estimates and confidence intervals using scatter plot glyphs.

A point plot represents an estimate of central tendency for a numeric variable by the position of scatter plot points and provides some indication of the uncertainty around that estimate using error bars.

```python
print("point")
sns.catplot(x="day", y="total_bill", kind="point", data=tips)
```

```python
print("bar")
sns.catplot(x="day", y="total_bill", kind="bar", data=tips) 
```

```python
print("count")
sns.catplot(x="day", kind="count", data=tips, palette="ch:.25", hue='size',col='sex',row='smoker') 
#kind="strip","swarm","box","violin","boxen","point","bar","count"
```
