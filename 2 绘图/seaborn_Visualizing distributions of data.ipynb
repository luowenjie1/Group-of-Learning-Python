```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from scipy import stats

import seaborn as sns
sns.set(style="ticks", color_codes=True)
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
penguins.info()
```

# seaborn.histplot

common_normbool

If True and using a normalized statistic, the normalization will apply over the full dataset. Otherwise, normalize each histogram independently.

```python
sns.displot(penguins,x="flipper_length_mm",binwidth=3,kde=True,rug=True)
sns.displot(penguins,x="flipper_length_mm",binwidth=3,kde=True,rug=True,stat="frequency")
sns.displot(data=penguins,x="flipper_length_mm",hue='island',kind='hist',kde=True,rug=True,stat="probability", common_norm=True, multiple="stack")
sns.displot(data=penguins,x="flipper_length_mm",hue='species',kind='hist',kde=True,rug=True,stat="density", common_norm=False, multiple="stack")
```

# seaborn.ecdfplot

Plot empirical cumulative distribution functions.

```python
sns.displot(data=penguins, x="flipper_length_mm",kind='ecdf',rug=True, hue='island')
```

# seaborn.kdeplot

Plot univariate or bivariate distributions using kernel density estimation.

```python
sns.displot(data=penguins, x="flipper_length_mm",kind='kde',rug=True,hue='species',fill=True)
```

```python
sns.displot(penguins,x="flipper_length_mm",y="bill_depth_mm",hue='species', rug=True, kind='kde',log_scale=(False, False))
```

```python
sns.displot(penguins,x="species",y="island",rug=True)
```

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm",hue='species')
```

```python
sns.displot(
    penguins, x="bill_length_mm", y="bill_depth_mm",
    hue='species', rug=True)
```

```python
sns.pairplot(penguins,hue='species')
```

```python
sns.pairplot(data=penguins, diag_kind='hist', kind='reg', hue='species')
```
